```python
# Distribution plot of blood-pressure column
sns.displot(data.BloodPressure)
```

```python
#Replacing 0 with mean values in BloodPressure column as it is a Symmetric Plot
data['BloodPressure']=data['BloodPressure'].replace(0,data['BloodPressure'].mean())
```

```python
data.head(10)
```

```python
# Displaying Distribution plot of Insulin
sns.distplot(data.Insulin)
```

```python
# As it a right-skewed plot , hence we will replace 0 with median values
data['Insulin']=data['Insulin'].replace(0,data['Insulin'].median())
```

```python
# Displaying top-10 values of the dataset
data.head(10)
```

```python
# Displaying distribution plot of SkinThickness
sns.distplot(data.SkinThickness)
```

```python
# As it is somewhat symmetric plot , hence we will replace 0 with mean values
data['SkinThickness']=data['SkinThickness'].replace(0,data['SkinThickness'].mean())
```

```python
# Displaying top 10 values in a dataset
data.head(10)
```

```python
# Data imbalancing
data.Outcome.value_counts()
```

```python
# Statistical Measures of the data
data.describe()
```

```python
# To detect Outliers - we plot Boxplot
fig , ax =plt.subplots(figsize=(15,15))
sns.boxplot(data=data,ax=ax)
plt.savefig('boxPlot.png')
```

```python
#Removing Outliers
q=data['Insulin'].quantile(0.95)
```

```python
newData=data[data['Insulin']<q]
newData.shape
```

```python
#Again plotting the Boxplot to check for Outliers
fig , ax =plt.subplots(figsize=(15,15))
sns.boxplot(data=newData,ax=ax)
plt.savefig('newBoxPlot.png')
```

```python
#Split the features and the target column
x=newData.drop(columns='Outcome')
y=newData['Outcome']
```
